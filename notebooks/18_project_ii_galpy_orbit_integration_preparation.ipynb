# Project II — galpy Orbit Integration Preparation

This notebook prepares the candidate table for future orbit integration with `galpy`.

It checks whether the required 6D phase-space inputs are available and summarizes the readiness of the candidate sample.


```python
import pandas as pd

path = '../data/processed/project_ii_orbital_family_candidates.csv'
df = pd.read_csv(path)

print(df.shape)
df.head()
```

```python
required_orbit_columns = [
    'source_id',
    'ra_gaia',
    'dec_gaia',
    'pmra_recovered',
    'pmdec_recovered',
    'rv',
    'distance_for_l_kpc',
    'distance_recovery_source',
    'distance_quality_flag',
    'feh',
    'orbital_family_interpretation',
    'rotation_class',
    'inclination_proxy_class',
]

audit = pd.DataFrame({
    'column': required_orbit_columns,
    'available': [c in df.columns for c in required_orbit_columns],
    'non_null_count': [df[c].notna().sum() if c in df.columns else 0 for c in required_orbit_columns],
    'total_rows': len(df),
})
audit['non_null_fraction'] = audit['non_null_count'] / audit['total_rows']
audit
```

```python
ready_mask = (
    df['ra_gaia'].notna()
    & df['dec_gaia'].notna()
    & df['pmra_recovered'].notna()
    & df['pmdec_recovered'].notna()
    & df['rv'].notna()
    & df['distance_for_l_kpc'].notna()
)

print('total:', len(df))
print('ready:', int(ready_mask.sum()))
print('not ready:', int((~ready_mask).sum()))
```
